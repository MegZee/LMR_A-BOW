# Attempting to calculate time

```python
noob_stamps = []
for index, i in lBig_noob:
  noob_stamps.append(i['timestamp'])

list_duration_noob = []

for i in noob_stamps:
  df_temp = pd.DataFrame(i)
  calc = df_temp.iloc[len(df_temp)-1] - df_temp.iloc[0]
  list_duration_noob.append(calc.mean()/60000)

print(list_duration_noob)
print(len(list_duration_noob))
avg_noob = sum(list_duration_noob)/len(list_duration_noob)
print(avg_noob//1)
```

```python
pro_stamps = []
for index, i in lBig_pro:
  pro_stamps.append(i['timestamp'])

list_duration_pro = []

for i in pro_stamps:
  df_temp = pd.DataFrame(i)
  calc = df_temp.iloc[len(df_temp)-1] - df_temp.iloc[0]
  list_duration_pro.append(calc.mean()/60000)

print(list_duration_pro)
print(len(list_duration_pro))
avg_pro = sum(list_duration_pro)/len(list_duration_pro)
print(avg_pro//1)
```
